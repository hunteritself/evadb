### Create a class

```python
def create_weaviate_class(class_name: str, vectorizer: str, module_config: dict, properties: list) -> None:
    # In Weaviate, vector index creation and management is not explicitly done like Pinecone
    # Need to typically define a property in the schema to hold vectors and insert data accordingly

    """
     Create a Weaviate class with the specified configuration.

     Args:
         class_name (str): The name of the class to create, e.g., "Article".
         vectorizer (str): The vectorizer module to use, e.g., "text2vec-cohere".
         module_config (dict): Configuration for vectorizer and generative module, e.g.,
             {
                 "text2vec-cohere": {
                     "model": "embed-multilingual-v2.0",
                 },
             }
         properties (list): List of dictionaries specifying class properties, e.g.,
             [
                 {
                     "name": "title",
                     "dataType": ["text"]
                 },
                 {
                     "name": "body",
                     "dataType": ["text"]
                 },
             ]

     Returns:
         None
     """
    # Check if the class already exists
    if client.schema.exists(class_name):
        client.schema.delete_class(class_name)

    # Define the class object with provided parameters
    class_obj = {
        "class": class_name,
        "vectorizer": vectorizer,
        "moduleConfig": module_config,
        "properties": properties
    }


    # Call the Weaviate API to create the class
    client.schema.create_class(class_obj)

    # response = client.schema.get(class_name)
    # Check the response for success or handle any errors
    if client.schema.get(class_name)['class'] == class_name:
        print(f"Successfully created Weaviate class '{class_name}'")
    else:
        print(f"Failed to create Weaviate class '{class_name}'")

    return None
```

### Delete a class

```python
def delete_weaviate_class(class_name: str) -> None:
    """
    Delete a Weaviate class and its data.

    Args:
        class_name (str): The name of the Weaviate class to delete.

    Returns:
        None
    """
    # Call the Weaviate API to delete the class
    client.schema.delete_class(class_name)

    try:
        # Attempt to retrieve the class, and if it results in an exception,
        # consider the class as successfully deleted.
        client.schema.get(class_name)
        print(f"Failed to delete Weaviate class '{class_name}'")
    except Exception as e:
        print(f"Successfully deleted Weaviate class '{class_name}'")

    return None
```

```python
from typing import List

def add_to_weaviate_class(class_name: str, data_objects: List[dict]) -> None:
    """
    Add objects to the specified Weaviate class.

    Args:
        class_name (str): The name of the Weaviate class to add objects to.
        data_objects (List[dict]): A list of dictionaries, where each dictionary contains property names and values.

    Returns:
        None
    """
    # Iterate over each data object and add it to the Weaviate class
    for data_object in data_objects:
        client.data_object.create(data_object, class_name)

    return None
```

### Queries

```python
from dataclasses import dataclass

@dataclass
class VectorIndexQuery:
    embedding: list
    top_k: int


def query_weaviate_class(class_name, properties_to_retrieve, query: VectorIndexQuery) -> List[dict]:
    """
    Perform a similarity-based search in Weaviate.

    Args:
        class_name (str): The name of the Weaviate class to perform the search on.
        properties_to_retrieve (List[str]): A list of property names to retrieve.
        query (VectorIndexQuery): A query object for similarity search, containing the query vector and top_k.

    Returns:
        List[dict]: A list of dictionaries containing the retrieved properties.
    """
    try:
        # Define the similarity search query
        response = (
            client.query
            .get(class_name, properties_to_retrieve)
            .with_near_vector({
                "vector": query.embedding
            })
            .with_limit(query.top_k)
            .with_additional(["distance"])
            .do()
        )

        # Check if the response contains data
        data = response.get('data', {})
        if 'Get' not in data or class_name not in data['Get']:
            print(f"No objects of class {class_name} found.")
            return []

        # Extract the results
        results = data['Get'][class_name]
        print(results)

        return results

    except Exception as e:
        print(f"Failed to query Weaviate class '{class_name}'")
        print(e)

        return []
```

```python
def add_to_weaviate_class_with_uuid(class_name: str, data_objects: List[dict]) -> None:
    """
    Add objects with specified UUIDs to the Weaviate class.

    Args:
        class_name (str): The name of the Weaviate class to add objects to.
        data_objects (List[dict]): A list of dictionaries, where each dictionary contains property names, values, and a UUID.

    Returns:
        None
    """
    for data_object in data_objects:
        uuid = data_object.get("uuid")
        if uuid:
            # Remove the UUID from the properties dictionary before creating the object
            properties = {k: v for k, v in data_object.items() if k != "uuid"}
            client.data_object.create(data_object=properties, class_name=class_name, uuid=uuid)
        else:
            # If no UUID is provided, let Weaviate generate one
            client.data_object.create(data_object=data_object, class_name=class_name)

    return None

class_name = "Question"
data_objects = [
    {
        "uuid": "123e4567-e89b-12d3-a456-426614174000",
        "question": "What is the capital of France?",
        "answer": "Paris"
    }
]

add_to_weaviate_class_with_uuid(class_name, data_objects)
```

#### Update specific properties of an object

```python
def update_object_properties(class_name: str, uuid: str, properties: dict) -> None:
    """
    Update specific properties of an object in Weaviate.

    Args:
        class_name (str): The class name of the object to update.
        uuid (str): The UUID of the object to update.
        properties (dict): A dictionary of properties to update.
    """
    client.data_object.update(
        data_object=properties,
        class_name=class_name,
        uuid=uuid
    )

def test_update_object_properties():
    class_name = "Question"
    uuid = "123e4567-e89b-12d3-a456-426614174000"  # Replace with the actual UUID of the object to update
    properties = {"answer": "New York"}

    update_object_properties(class_name, uuid, properties)
    print(f"Updated properties for object {uuid} in class {class_name}")

# Run the test
test_update_object_properties()
```

#### Replace an object in its entirety

```python
def replace_object(class_name: str, uuid: str, properties: dict) -> None:
    """
    Replace an object in its entirety in Weaviate.

    Args:
        class_name (str): The class name of the object to replace.
        uuid (str): The UUID of the object to replace.
        properties (dict): A dictionary of the new properties of the object.
    """
    client.data_object.replace(
        data_object=properties,
        class_name=class_name,
        uuid=uuid
    )

def test_replace_object():
    class_name = "Question"
    uuid = "123e4567-e89b-12d3-a456-426614174000"  # Replace with the actual UUID of the object to replace
    properties = {"question": "New question", "answer": "Atlanta", "category": "New category"}

    replace_object(class_name, uuid, properties)
    print(f"Replaced object {uuid} in class {class_name}")

# Run the test
test_replace_object()
```

#### Delete specific properties from an object

```python
def delete_object_properties(class_name: str, uuid: str, properties_to_delete: List[str]) -> None:
    """
    Delete specific properties from an object in Weaviate

    Args:
        class_name (str): The class name of the object to update.
        uuid (str): The UUID of the object to update.
        properties_to_delete (List[str]): A list of property names to delete.
    """

    # Fetch the current object data
    object_data = client.data_object.get_by_id(uuid)['properties']

    # Remove the properties to delete
    updated_properties = {
        key: value for key, value in object_data.items() if key not in properties_to_delete
    }

    # Replace the object with the updated properties
    client.data_object.replace(
        data_object=updated_properties,
        class_name=class_name,
        uuid=uuid
    )

def test_delete_object_properties():
    class_name = "Question"
    uuid = "123e4567-e89b-12d3-a456-426614174000"
    properties_to_delete = ["category"]

    delete_object_properties(class_name, uuid, properties_to_delete)
    print(f"Deleted properties {properties_to_delete} from object {uuid} in class {class_name}")

test_delete_object_properties()
```

```python
def keyword_search(class_name: str, query: str, properties: List[str], limit: int = 10, autocut: int = None, properties_to_retrieve: List[str] = None):
    """
    Perform a BM25 keyword search in Weaviate.

    Args:
        class_name (str): The name of the Weaviate class to search in.
        query (str): The keyword query string.
        properties (List[str]): The properties to consider in the search.
        limit (int, optional): The maximum number of results to return. Defaults to 10.
        autocut (int, optional): The number of significant breaks in score to use for grouping results. If None, all results are returned.
        properties_to_retrieve (List[str], optional): The properties to retrieve from the search results. If None, all properties are retrieved.

    Returns:
        List[dict]: A list of dictionaries containing the search results.
    """

    try:
        # Build the BM25 search query
        search_query = (
            client.query
            .get(class_name, properties_to_retrieve or ["*"])
            .with_bm25(query=query, properties=properties)
        )

        # Apply autocut if provided
        if autocut is not None:
            search_query = search_query.with_autocut(autocut)

        # Set the result limit and execute the query
        response = search_query.with_limit(limit).do()

        # Check if the response contains data
        data = response.get('data', {})
        if 'Get' not in data or class_name not in data['Get']:
            print(f"No objects of class {class_name} found for query '{query}'.")
            return []

        # Extract the results
        results = data['Get'][class_name]
        return results

    except Exception as e:
        print(f"Failed to perform keyword search in class '{class_name}'")
        print(e)
        return []


def test_keyword_search():
    class_name = "Question"
    query_string = "animal"
    search_properties = ["question", "answer"]
    limit = 3
    properties_to_retrieve = ["question", "answer"]

    results = keyword_search(class_name, query_string, search_properties, limit, properties_to_retrieve=properties_to_retrieve)

    print("Search Results:")
    for result in results:
        print(result)

test_keyword_search()
```

#### Hybrid search

```python
def hybrid_search(class_name: str, query: str, properties: List[str] = None, limit: int = 10):
    """
    Perform a hybrid search in Weaviate, combining keyword and vector search.

    Args:
        class_name (str): The name of the Weaviate class to search in.
        query (str): The keyword query string.
        vector (List[float], optional): The vector for vector search. If None, only BM25 search is performed.
        properties (List[str], optional): The properties to consider in the BM25 search.
        limit (int, optional): The maximum number of results to return. Defaults to 10.
        alpha (float, optional): The balance between keyword and vector search. Defaults to 0.5.
        autocut (int, optional): The number of significant breaks in score for autocut. If None, all results are returned.
        additional (List[str], optional): Additional properties to retrieve in the search results.

    Returns:
        List[dict]: A list of dictionaries containing the search results.
    """

    try:
        response = (
            client.query
            .get(class_name, properties)
            .with_hybrid(
                query=query
            )
            .with_limit(limit)
            .with_additional(["distance"])
            .do()
        )

        # Check if the response contains data
        data = response.get('data', {})
        if 'Get' not in data or class_name not in data['Get']:
            print(f"No objects of class {class_name} found for query '{query}'.")
            return []

        # Extract the results
        results = data['Get'][class_name]
        return results

    except Exception as e:
        print(f"Failed to perform hybrid search in class '{class_name}'")
        print(e)
        return []
        
def hybrid_search_test():
    class_name = "Question"
    query_string = "animal"
    search_properties = ["question", "answer"]
    limit = 3
    additional_properties = ["question", "answer", "round"]

    results = hybrid_search(class_name, query=query_string, properties=search_properties, limit=limit)
    
    print("Hybrid Search Results:")
    for result in results:
        print(result)

hybrid_search_test()
```

#### Generative search

```python
def generative_search(
    class_name: str, 
    query: str, 
    prompt: str, 
    properties: List[str] = None, 
    limit: int = 10, 
    grouped_task: bool = False, 
    grouped_properties: List[str] = None
):
    """
    Perform a generative search in Weaviate using a large language model.

    Args:
        class_name (str): The name of the Weaviate class to search in.
        query (str): The keyword query string for initial object retrieval.
        prompt (str): The prompt to use for generating text with the LLM.
        properties (List[str], optional): The properties to consider in the BM25 search.
        limit (int, optional): The maximum number of results to return. Defaults to 10.
        grouped_task (bool, optional): Whether to perform a grouped task. Defaults to False.
        grouped_properties (List[str], optional): Properties to include in the prompt for a grouped task.

    Returns:
        List[dict]: A list of dictionaries containing the search and generated results.
    """
    try:
        # Build the generative search query
        search_query = (
            client.query
            .get(class_name, properties)
            .with_near_text({"concepts": [query]})
        )

        # Apply generative search based on the specified prompt
        if grouped_task:
            search_query = search_query.with_generate(
                grouped_task=prompt, 
                grouped_properties=grouped_properties
            )
        else:
            search_query = search_query.with_generate(single_prompt=prompt)

        # Set the result limit and execute the query
        response = search_query.with_limit(limit).do()

        # Check if the response contains data
        data = response.get('data', {})
        if 'Get' not in data or class_name not in data['Get']:
            print(f"No objects of class {class_name} found for query '{query}'.")
            return []

        # Extract the results
        results = data['Get'][class_name]
        return results

    except Exception as e:
        print(f"Failed to perform generative search in class '{class_name}'")
        print(e)
        return []
    
    
def generative_search_test():
    class_name = "Question"
    query_string = "animal"
    prompt = "Convert the following into a question for twitter. Include emojis for fun, but do not include the answer: {question}."
    limit = 2
    properties = ["question"]

    results = generative_search(class_name, query=query_string, prompt=prompt, properties=properties, limit=limit)

    print("Generative Search Results:")
    for result in results:
        print("Question:", result['question'])
        if '_additional' in result and 'generate' in result['_additional']:
            generate_result = result['_additional']['generate']
            if 'singleResult' in generate_result:
                print("Generated Response:", generate_result['singleResult'])
            elif 'groupedResult' in generate_result:
                print("Grouped Generated Response:", generate_result['groupedResult'])
            elif 'error' in generate_result:
                print("Error:", generate_result['error'])
            else:
                print("No generated response found.")
        else:
            print("No additional generated information.")

generative_search_test()
```
